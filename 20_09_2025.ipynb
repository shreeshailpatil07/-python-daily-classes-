```python
# compile time / method overloading polymorphism

class BCA_1st_Semester :
  def student(self,a,b=0,c=0,d=0):
      return a+b+c+d

BCA = BCA_1st_Semester()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,4,-12))
```
